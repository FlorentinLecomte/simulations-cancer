```matlab
clearvars
close all
x = linspace(0,1,100);
t = 0:20;
m = 0;
sol = pdepe(m,@modelpde,@modelic,@modelbc,x,t);
```

```matlab
n = sol(:,:,1);
f = sol(:,:,2);
m = sol(:,:,3);
```

```matlab
surf(x,t,n)
title('n(x,t): Concentration of cancer cells')
xlabel('Distance x')
ylabel('Time t')
saveas(gcf, 'surfn.png');
```

```matlab
surf(x,t,f)
title('f(x,t): Concentration of cancer cells')
xlabel('Distance x')
ylabel('Time t')
saveas(gcf, 'surff.png');
```

```matlab
surf(x,t,m)
title('m(x,t): Concentration of cancer cells')
xlabel('Distance x')
ylabel('Time t')
saveas(gcf, 'surfm.png');
```

```matlab
plot(x,n(1,:))
hold on;
plot(x,f(1,:))
plot(x,m(1,:))
title('t=0')
xlabel('x');
legend('Cancer cells', 'ECM', 'MDE', 'Location', 'northeast');
hold off;
saveas(gcf, 'plot_t0.png');
```

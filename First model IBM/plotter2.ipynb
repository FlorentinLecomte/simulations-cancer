# Plotting the solution of the discrete model

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Retrieve the data

```python
data = np.fromfile('simulation_data.bin', dtype=np.float64)
df = data.reshape(-1, 3, 202, 202)
```

### Using 'Imshow'

```python
vmax_n = np.max(df[:, 0, 1:-1, 1:-1])

for t_idx in range(df.shape[0]):
    grid = df[t_idx, 0, 1:-1, 1:-1]

    fig, ax = plt.subplots()
    im = ax.imshow(grid, cmap='Reds', origin='lower', extent=[0, 1, 0, 1], vmin=0, vmax=vmax_n)
    # plt.colorbar(im, ax=ax, label="Density of cells")
    ax.set_title(f"t={t_idx*0.1:0.1f}")
    ax.set_xlabel("x")
    ax.set_ylabel("y")

    filename = f"figures/frame_{t_idx:03d}.png"
    # plt.savefig(filename, dpi=200, bbox_inches='tight', pad_inches=0.1)
    plt.savefig(filename)

    plt.close(fig)
```

### Using 'Scatter'

```python
for t_idx in range(df.shape[0]):
    grid = df[t_idx, 0, 1:-1, 1:-1]

    y, x = np.where(grid > 0)

    fig, ax = plt.subplots(figsize=(6, 6))

    ax.scatter(x/200, y/200, s=2, c='black')

    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)

    ax.set_title(f"t={t_idx*0.1:0.1f}")
    ax.set_xlabel("x")
    ax.set_ylabel("y")

    filename = f"figures/frame_{t_idx:03d}.png"
    plt.savefig(filename)
    plt.close(fig)
```

### Creating a video

```python
import imageio.v2 as imageio
import glob

images = sorted(glob.glob("figures/frame_*.png"))
frames = [imageio.imread(f) for f in images]

imageio.mimsave('animations/animation.mp4', frames, fps=30)
```
